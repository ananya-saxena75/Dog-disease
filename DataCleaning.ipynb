DATA CLEANING

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Stray dog survey .csv")
```

```python
df.head()
```

Remove Unnecessary Columns

```python
df = df.drop(columns=["Timestamp", "Email Address", "Name"], errors='ignore')
```

Check Missing Values

```python
print(df.isnull().sum())
```

```python
df = df.fillna(method="ffill")
```

Standardize Text Values

```python
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
df.replace({
    "yes": 1,
    "no": 0
}, inplace=True)
```

Encode Severity Column

```python
df.rename(columns={
    "Rate the severity of the stray dog issue in your area": "severity"
}, inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

Remove Duplicate Rows

```python
df = df.drop_duplicates()
```
